```python
X=data[['x','y']]
y=data['sum']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from tkinter import *
import joblib

master=Tk()
master.title("Addition of two numbers using ML")
label=Label(master,text="Addtion of two numbers uding ML",bg='green',fg='white').grid(row=0,columnspan=2)
```

```python
def show_entry_fields():
    p1=float(e1.get())
    p2=float(e2.get())

    model=joblib.load('finalmodel')
    result=model.predict([[p1,p2]])

    Label(master,text='sum is=').grid(row=4)
    Label(master,text=result).grid(rows=5)

Label(master,text="Enter First Number").grid(row=1)
Label(master,text="Enter Second Number").grid(row=2)

e1=Entry(master)
e2=Entry(master)

e1.grid(row=1,column=1)
e2.grid(row=2,column=1)
Button(master,text='predict',command=show_entry_fields).grid()

mainloop()


```
